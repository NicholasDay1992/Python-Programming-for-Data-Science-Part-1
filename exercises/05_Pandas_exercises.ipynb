# Using the BMI data in the Series

We can also utilise the numpy arrays we set up last week for the data and labels of the BMI dataset.  

```python

# Method 1: Convert each row to a list and get the first 4 items
hw_data = np.array([list(row)[1:3] for row in hw_dataset])
hw_labels = np.array([list(row)[0] for row in hw_dataset])

#
hw_by_label = {}
for label in set(hw_labels):
    hw_by_label[label] = hw_data[hw_labels == label]
print(hw_by_label)
```

# Exercise X
Now calculate the mean weight and height for each gender and display in a dataframe.


```python
mean_height = bmi_df.groupby('Gender').mean()['Height']
mean_weight = bmi_df.groupby('Gender').mean()['Weight']

df = pd.DataFrame([mean_height, mean_weight])
df
```
